```python
#4. Calculate the following information using the merged dataset :
#a. Time period of the available transaction data
#b. Count of transactions where the total amount of transaction was negative
```

```python
#a. Time period of the available transaction data
Customer_Final['tran_date']= pd.to_datetime(Customer_Final['tran_date'], format = 'mixed')
min_date = Customer_Final['tran_date'].min()
max_date = Customer_Final['tran_date'].max()

time_period = max_date - min_date
print('Time period: ',time_period.days)
```

```python
Customer_Final['tran_date'].dtype
```

```python
#b. Count of transactions where the total amount of transaction was negative
negative_transaction = (Customer_Final['total_amt']<0).sum()
print('Count of transaction where the amount is negative: ',negative_transaction)
```

```python
#7. Which store type sells the maximum products by value and by quantity?
```

```python
max_quantity = Customer_Final.groupby('Store_type').agg({'Qty':np.max})
max_value = Customer_Final.groupby('Store_type').agg({'Rate':np.max})
```

```python
pd.concat([max_quantity,max_value], axis = 1).sort_values(by ='Rate', ascending = False ).reset_index()
```

```python
#8. What was the total amount earned from the "Electronics" and "Clothing" categories from 
#Flagship Stores?
```

```python
#9. What was the total amount earned from "Male" customers under the "Electronics" category?
```

```python
Customer_Final.pivot_table(index = ['Gender'],columns = ['prod_cat'],values = 'total_amt', aggfunc = {'total_amt':np.sum}).loc[['M'],['Electronics']]
```

```python
#10. How many customers have more than 10 unique transactions, after removing all transactions 
#which have any negative amounts?
```

```python
Customer_Final = Customer_Final[Customer_Final['total_amt']>0]
unique_transactions = Customer_Final.groupby('customer_Id')['transaction_id'].nunique()
dd = unique_transactions[unique_transactions > 10]
print('Customers have more than 10 unique transactions: ',len(dd))
```

```python
#11. For all customers aged between 25 - 35, find out:
#a. What was the total amount spent for “Electronics” and “Books” product categories?
#b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?
```

```python
Customer_Final['DOB'] = pd.to_datetime(Customer_Final['DOB'], format = '%d-%m-%Y')
```

```python
Customer_Final['Customers_age'] = (dt.datetime.today() - Customer_Final['DOB']).dt.days//365.25
Customer_Final['Customers_age']
```

```python
filtered = Customer_Final.loc[(Customer_Final['Customers_age']>=25) & (Customer_Final['Customers_age']<=35)] 
```

```python
#a. What was the total amount spent for “Electronics” and “Books” product categories?
```

```python
filtered.groupby(by = 'prod_cat').agg({'total_amt':np.sum}).loc[['Books','Electronics']].sum()
```

```python
#b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?
```

```python
filtered['tran_date'] = pd.to_datetime(filtered['tran_date'], format = '%Y-%m-%d')
```

```python
start_dt = '2014-01-01'
end_dt = '2014-03-01'
filtered.groupby(by = 'tran_date').agg({'total_amt':np.sum}).loc[start_dt:end_dt].sum()
```
